```python
token = getToken(client_id, client_secret)
```

```python
# request the playlist info via Spotify API and select the items parts
playlist_tracks = getPlaylist(token, new_year_playlist_id)['items']
#playlist_tracks
```

```python
len(playlist_tracks)
```

```python
#create empty dataframe
df_spotify = pd.DataFrame(columns=['name', 'artist', 'track_id'], index=range(len(playlist_tracks)))
```

```python
# populate the dataframe with json response
for idx,item in enumerate(playlist_tracks):
    name = item['track']['name']
    artist = item['track']['artists'][0]['name']
    track_id = item['track']['id']
    df_spotify.iloc[idx] = [name, artist, track_id]
```

```python
df_spotify
```

```python
# query Spotify API for audi features & analysis and add them as a column
df_spotify['audio_features'] = df_spotify.apply(lambda x: getFeature(x['track_id']), axis=1)
df_spotify['audio_analysis'] = df_spotify.apply(lambda x: getAnalysis(x['track_id']), axis=1)
df_spotify.head()
```

```python
# add selected song features as attribute
def getFeature(row, feature):
    return row['audio_features'][0][feature]

#add selected features as a column
features = ['energy', 'speechiness', 'valence', 'tempo', 'duration_ms']

for f in features:
    df_spotify[f] = df_spotify.apply(lambda x: getFeature(x,f), axis=1)

df_spotify.head()
```

```python
# add selected attributes of sections in a sections attribute
sectionsToDrop = ['confidence', 'tempo_confidence', 'key_confidence', 'mode', 'mode_confidence', 'time_signature', 'time_signature_confidence']

def getSimplifiedSections(row):
    sectionList = row['audio_analysis']['sections']
    for s in sectionList:
        for d in sectionsToDrop:
            if d in s:
                del s[d]
    
    return sectionList
    

df_spotify['sections'] = df_spotify.apply(lambda x: getSimplifiedSections(x), axis=1)

df_spotify.head()
```

```python
# keep only the desired attributes
df_final = df_spotify[['name', 'artist', 'track_id', 'energy', 'speechiness', 'valence', 'tempo', 'duration_ms', 'sections' ]]
df_final.head()
```

```python
# save as json
with open('ne.json', 'w') as f:
    f.write(df_final.to_json(orient='records'))
```
