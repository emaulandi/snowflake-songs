```python
#https://spotipy.readthedocs.io/en/latest/#examples
#https://github.com/plamere/spotipy

import json
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

import unicodecsv
import pandas as pd
from IPython.display import display, HTML
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# function read csv
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)

def typeDf(df,to_be_parse):
    for col in to_be_parse:
        if col['type'] == 'int':
            df[col['attribute']] = df[col['attribute']].astype(int)
        elif col['type'] == 'float':
            df[col['attribute']] = df[col['attribute']].astype(float)
```

```python
#load CSV as list
filename = 'christmas-song.csv'
alist = read_csv(filename)

len(alist)
```

```python
#for index, row in df_spotify.iterrows() :
#    features = sp.audio_features(tid_base + row['Spotify ID'])
df_spotify['audio_features'] = df_spotify.apply(lambda x: getFeature(x['Spotify ID']), axis=1)
df_spotify['audio_analysis'] = df_spotify.apply(lambda x: getAnalysis(x['Spotify ID']), axis=1)
```

```python
df_clean = df_spotify.reset_index()
```

```python
#df.drop('index', axis=1)
df_clean.drop('index', axis=1, inplace=True)
df_clean.drop('Song (Wiki + Google)', axis=1, inplace=True)

df_clean.head()
```

```python
#for index, row in df_clean.iterrows():
#    if row['audio_analysis'] == None:
#        df_clean.drop(df_clean.index[index], inplace=True)
#        df_clean.reset_index(inplace=True)
#        print('None')
df_clean2 = df_clean[(df_clean.Country != 'Chile') & (df_clean.Country != 'Australia')]

print(len(df_clean))
print(len(df_clean2))
```

```python
def getFeature(row, feature):
    return row['audio_features'][0][feature]

#add selected features as a column
features = ['energy', 'speechiness', 'valence', 'tempo', 'duration_ms']

for f in features:
    df_clean2[f] = df_clean2.apply(lambda x: getFeature(x,f), axis=1)

df_clean2.head()
```

```python
segmentsToDrop = ['confidence', 'loudness_start', 'loudness_max', 'loudness_max_time', 'loudness_end']
sectionsToDrop = ['confidence', 'tempo_confidence', 'key_confidence', 'mode', 'mode_confidence', 'time_signature', 'time_signature_confidence']

def getSimplifiedSegments(row):
    segmentList = row['audio_analysis']['segments']
    for s in segmentList:
        for d in segmentsToDrop:
            if d in s:
                del s[d]
    
    return segmentList

def getSimplifiedSections(row):
    sectionList = row['audio_analysis']['sections']
    for s in sectionList:
        for d in sectionsToDrop:
            if d in s:
                del s[d]
    
    return sectionList
    

df_clean2['segments'] = df_clean2.apply(lambda x: getSimplifiedSegments(x), axis=1)
df_clean2['sections'] = df_clean2.apply(lambda x: getSimplifiedSections(x), axis=1)

df_clean2.head()
```

```python
df_final = df_clean2[['Country','Population','SongTitle', 'Spotify ID', 'energy', 'speechiness', 'valence', 'tempo', 'duration_ms', 'sections' ]]

df_final.head()
```

```python
with open('temp.json', 'w') as f:
    f.write(df_final.to_json(orient='records', lines=True))
```
